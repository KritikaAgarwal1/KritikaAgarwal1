Lets split the data into training and testin

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)
```

As we know , we have to select the best regression model which is best suitable for this dataset , so for the best performer will do some hyper paramter tunning 

```python
param={
    'linear':{
        'model': LinearRegression(),
        'param':{}},
    'Ridge':{
        'model': Ridge(),
        'param':{
            'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
            'alpha':[0,10,100]}},
    'Lasso':{
        'model':Lasso(),
        'param':{
            'alpha':[0,0.1,1]}},
    'RFR':{
        'model': RandomForestRegressor(),
        'param':{
            'n_estimators':[1,10,100,1000],
            'criterion':["squared_error", "absolute_error", "friedman_mse", "poisson"]}},
    'svr':{
        'model': SVR(),
        'param':{
            'kernel':['linear','rbf','sigmoid'],
            'C':[1,5,10],
            'epsilon':[0.1,0.001,0.0001]}}
}
```

```python
score = []
for model_name , mp in param.items():
    clf=GridSearchCV(mp['model'],mp['param'],cv=5,return_train_score = False)
    clf.fit(X_train_pca , y_train)
    score.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_prams' : clf.best_params_
    })
```

```python
dff = pd.DataFrame(score,columns=['model','best_score','best_prams'])
dff
```

Now lets see the model scores 

```python
def scores(true, predicted):  
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    print('MODEL EVALUATION ------------       ')
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
```
