```python
new_df = pd.get_dummies(df, columns=['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet',
 'romantic'])
```

```python
X = new_df.drop('G3', axis=1).values
y = new_df['G3'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('std_scalar', StandardScaler())
])

X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
X_test
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(0.95)
X_pca = pca.fit_transform(X)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(criterion='friedman_mse',n_estimators=1000)
model.fit(X_train_pca, y_train)
```

```python
model.score(X_test_pca,y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
param={
    'linear':{
        'model': LinearRegression(),
        'param':{}},
    'Ridge':{
        'model': Ridge(),
        'param':{
            'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs'],
            'alpha':[0,10,100]}},
    'Lasso':{
        'model':Lasso(),
        'param':{
            'alpha':[0,0.1,1]}},
    'RFR':{
        'model': RandomForestRegressor(),
        'param':{
            'n_estimators':[1,10,100,1000],
            'criterion':["squared_error", "absolute_error", "friedman_mse", "poisson"]}},
    'svr':{
        'model': SVR(),
        'param':{
            'kernel':['linear','rbf','sigmoid'],
            'C':[1,5,10],
            'epsilon':[0.1,0.001,0.0001]}}
}
```

```python
score = []
for model_name , mp in param.items():
    clf=GridSearchCV(mp['model'],mp['param'],cv=5,return_train_score = False)
    clf.fit(X_train_pca , y_train)
    score.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_prams' : clf.best_params_
    })
```

```python
dff = pd.DataFrame(score,columns=['model','best_score','best_prams'])
dff
```
